```python
#Import Packages
import numpy as np
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
##Let's take a look at the full text of the Game of Thrones Series, Book 1

#Let's set up and read the file in. I am creating two different reading variables. text_file uses the readlines() 
#function so that we can show the first 30 lines of the text, where book is the entire book
filepath = r'S:\Data\GOT_Books\book1.txt' 
print('Reading in data...')
text_file = open(filepath,'r')
book = open(filepath,'r').read() # We'll save this variable for our actual cloud

n = 30
#Read the lines
text = text_file.readlines()
print("The following are the first",n,"lines of a text file:")
#Iterate the list of lines to retrieve the first N lines of a file
for line in (text[:n]):
# Printing the first 30 lines of the file line by line.
    print(line, end ='')
```

```python
#We're going to add a logo form for the wordcloud for added effect
background_image = np.array(Image.open(r'S:\Data\GOT_Books\GOT_logo.jpg'))
plt.figure(figsize=(15,15))
plt.imshow(background_image)
```

```python
#Creating word cloud with text as argument in .generate() method
word_cloud = WordCloud(background_color = 'white',mask = background_image, 
               width = 2048, height = 1080).generate(book)

#Getting the font color to match the image
img_colors = ImageColorGenerator(background_image)
word_cloud.recolor(color_func = img_colors)
```

```python
#Final Plot
plt.figure(figsize=(15,15))
plt.imshow(word_cloud, interpolation='bilinear') 
plt.axis("off")
plt.show()
```
